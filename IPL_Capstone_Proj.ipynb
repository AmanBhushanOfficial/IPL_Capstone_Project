**Here comes some Basic Questions**

*Who Won the most matches ??*

```python
match_wins=df['match_winner'].value_counts()
sns.barplot(y=match_wins.index, x= match_wins.values,palette='magma')
plt.title('Most match win by Team')
plt.xlabel('Win Count')
plt.ylabel('Team Names')
```

*2. Toss decision Trends*

```python
sns.countplot(x=df['toss_decision'],palette='Blues')
plt.title('Toss decision Trends')
```

*3. Toss Winner vs Match Winner*

```python
counti=df[df['toss_winner'] == df['match_winner']]['match_id'].count()
counti
```

```python
percentage = (counti*100)/df.shape[0]
print("percentage =",percentage.round(3))
```

*3. 10 Best Bowling Figures* 

```python
df.head()
```

```python
df['highest_wicket'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wicket']=df['highest_wicket'].astype(int)
```

```python
# df.groupby('highest_wicket')['best_bowling'].sum().sort_values(ascending=False).head(10)
```

```python
top_bowler=df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending=False).head(10).head(10)
top_bowler
```

```python
top_bowler.plot(kind='barh')
```
